```python
MODEL_PATH = "E:\\3DGS\\trained_models\\playroom-11-14-24\\point_cloud\\iteration_30000\\point_cloud.ply"
OUTPUT_PATH = "E:\\3DGS\\trained_models\\playroom-11-14-24\\point_cloud\\iteration_30000\\point_cloud_shuffled.ply"
```

```python
import numpy as np
from plyfile import PlyData, PlyElement

plydata = PlyData.read(MODEL_PATH)
plydata.elements
```

```python
new_plydata_elements = []
for element in plydata.elements:
    if element.name == "vertex":
        arr = element.data
        perm = np.random.permutation(len(arr))
        shuffled = arr[perm].copy()
        new_element = PlyElement.describe(shuffled, "vertex")
        new_plydata_elements.append(new_element)
    else:
        new_plydata_elements.append(element)

new_plydata = PlyData(new_plydata_elements, text=plydata.text, comments=plydata.comments)
new_plydata.write(OUTPUT_PATH)
```
