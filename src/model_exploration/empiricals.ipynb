```python
from pandas import read_csv
from os.path import join
from matplotlib.pyplot import figure, hist, xlabel, ylabel, title, tight_layout, show

METRICS_DIR = "metrics"
```

```python
def compute_metrics_and_histogram(filename: str):
    df = read_csv(join(METRICS_DIR, filename + ".csv"), header=None)
    col = df[0]
    print(f"Min:\n{col.min():,}\n")
    print(f"Max:\n{col.max():,}\n")
    print(f"Median:\n{int(col.median()):,}\n")
    print(f"Mean:\n{int(round(col.mean())):,}\n")
    
    figure(figsize=(8, 5))
    hist(col.dropna(), bins='auto', edgecolor='black')
    xlabel('Buckets')
    ylabel('Frequency')
    title(f'Histogram of {filename}')
    tight_layout()
    show()
```

```python
compute_metrics_and_histogram("fg--garden--n_instances")
```
