```python
MODEL_PATH = "E:\\3DGS\\trained_models\\playroom-11-14-24\\point_cloud\\iteration_30000\\point_cloud.ply"
```

```python
import matplotlib.pyplot as plt
from plyfile import PlyData

plydata = PlyData.read(MODEL_PATH)
plydata.elements
```

```python
x = plydata['vertex'].data["x"]
plt.figure(figsize=(8, 4))
plt.hist(x, bins=50, color='C0', edgecolor='k', range=(-7, 7), log=False)
plt.xlabel('x')
plt.ylabel('count')
plt.title('Distribution of x coordinates in point cloud')
plt.grid(True)
plt.show()
```

```python
y = plydata['vertex'].data["y"]
plt.figure(figsize=(8, 4))
plt.hist(y, bins=50, color='C0', edgecolor='k', range=(-7, 7), log=False)
plt.xlabel('y')
plt.ylabel('count')
plt.title('Distribution of y coordinates in point cloud')
plt.grid(True)
plt.show()
```
